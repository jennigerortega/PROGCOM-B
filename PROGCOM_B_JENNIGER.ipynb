10. TIPO DE TRIANGULO

```python
lado1=int(input("ingrese el primer lado "))
lado2=int(input("ingrese el segundo lado "))
lado3=int(input("ingrese el tercer lado "))
if lado1==lado2 and lado2==lado3:
  print("el triangulo es equilatero")
else:
  if lado1==lado2 or lado1==lado3 or lado2==lado3:
    print("el triangulo es isoceles")
  else:
    print("el triangulo es escaleno")
```

11. IMC

```python
nombre=input("ingrese su nombre ")
peso=float(input("ingrese su peso "))
estatura=float(input("ingrese su estatura "))
imc=peso/(estatura**2)
if imc<18.5:
  print(nombre,"usted tiene bajo peso")
else:
  if imc>=18.5 and imc<=24.9:
    print(nombre,"usted tiene peso normal")
  else:
    if imc>=25 and imc<=29.9:
      print(nombre,"usted tiene sobrepeso")
    elif imc>=30:
      print(nombre,"usted tiene obesidad")

  print(nombre,"su imc es",imc)


```

12. CALCULADORA SIMPLE

```python
num1=int(input("ingrese el primer numero "))
num2=int(input("ingrese el segundo numero "))
operacion=input("ingrese la operacion que desea realizar ")
if operacion=="+":
  print("la suma de los numeros es",num1+num2)
else:
  if operacion=="-":
    print("la resta de los numeros es",num1-num2)
  else:
    if operacion=="*":
      print("la multiplicacion de los numeros es",num1*num2)
```
