¿Cómo se halla el indice de masa corporal de una persona?

```python
peso= float(input("ingrese su peso: "))
altura= float(input("ingrese su altura en metros: "))
imc= (peso/altura**2)
print("su indice de masa corporal es: ", imc)

if imc <18.0:
    print("usted esta bajo de peso")
elif  18.5 <= imc <25.0:
    print("usted esta en un peso normal")
elif 25.5 <= imc <30.0:
    print("usted esta en sobrepeso")
else:
    print("usted esta en obesidad")


```
