19. Crea un programa que muestre los primeros n números de la serie Fibonacci.

```python
n=int(input("Ingrese un numero: "))
a=0
b=1
print(a)
print(b)
for i in range(n-2):
    c=a+b
    print(c)
    a=b
    b=c
print("Fin del programa")

```

22. Crea un programa que calcule la suma de los dígitos de un número.

```python
numero=int(input("Ingrese un numero de dos digitos: "))
suma=0
while numero>0:
    digito=numero%10
    suma=suma+digito
    numero=numero//10
print(f"La suma de los digitos del numero es {suma}")
```

23. Escribe un programa que determine si un número es primo.

```python
numero=int(input("Ingrese un numero: "))
contador=0
if numero==0 or numero==1:
    print("El numero no es primo")
    contador=1
else:
    for i in range(2,numero):
        if numero%i==0:
            print("El numero no es primo")
            contador=1
            break
if contador==0:
    print("El numero es primo")
```
